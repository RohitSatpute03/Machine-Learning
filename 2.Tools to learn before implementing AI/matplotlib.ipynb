```python
import matplotlib.pyplot as plt
import math as m
import numpy as np
```

```python
x = [2,3,4,5,6]
y = [1,2,3,4,5]

plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

```python
x = [2,3,4,5,6]
y = [1,2,3,4,5]

x1 = [1,3,4,5,6]
y1 = [7,6,5,4,3]


plt.scatter(x,y)
plt.plot(x,y, "b")
plt.scatter(x1,y1)
plt.plot(x1,y1,'y--')
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = 2*(x**4) + 100
plt.plot(x,y)
plt.show()
```

```python
x = np.arange( 4 ,32, 0.1)
y = 2*(x**4) + 100
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(3, 78, 0.2)
y = x ** 4
y2 = x ** 3

plt.plot(x,y, color='red',label='x^3')
plt.plot(x,y2, color='green', label='x^4')
plt.ylabel('Y', fontsize=15)
plt.xlabel('x', fontsize=15)
plt.legend()
plt.title("Prepinsta")
plt.text(10, 15, "Rohit IAS", fontsize="20")
plt.grid()
plt.show()
```

```python
profit = [10, 25, 40, 30]
labels= ['amazon', 'flipkart', 'Rohs', 'ola']
colors = ['red', 'blue', 'magenta', 'green']
explode = [0.2, 0.1,0,0]
plt.pie(profit, colors=colors, labels=labels, startangle=90, explode=explode)
plt.show()
```
