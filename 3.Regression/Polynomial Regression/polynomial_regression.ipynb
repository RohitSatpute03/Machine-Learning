Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing data sets


```python
data_set = pd.read_csv("/content/Dataset_master.xlsx - RBC in Humans Polynomial Regres.csv")
#age of the person and RBC's according to their age

x = data_set.iloc[:, :-1].values
y = data_set.iloc[:,-1].values
```

```python
print(x)
print(y)
```

training using Linear regression


```python
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()
linear_regressor.fit(x,y)
```

training using Polynomial regression


```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 6)
x_poly = poly_reg.fit_transform(x)


#Whenever we use polynimal regression we need to use it with linear regression always
linear_reg2 = LinearRegression()
linear_reg2.fit(x_poly, y)
```

  Visualizing linear regression result

```python
plt.scatter(x, y, color='red')
plt.title("RBC - Linear regression")
plt.plot(x, linear_regressor.predict(x))
plt.xlabel('age')
plt.ylabel('RBC')
plt.show()
```

Visualizing polynimial regression result

```python
plt.scatter(x, y, color='red')
plt.title("RBC - Polynomial regression")
plt.plot(x, linear_reg2.predict(x_poly))
plt.xlabel('age')
plt.ylabel('RBC')
plt.show()
```

predict a new result using linear regression

```python
linear_regressor.predict([[6.5]])
```

predict a new result using linear regression

```python
linear_reg2.predict(poly_reg.fit_transform([[6.5]]))
```
