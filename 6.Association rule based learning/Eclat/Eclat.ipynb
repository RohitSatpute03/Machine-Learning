Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing dataset

```python
data_set = pd.read_csv("/content/Dataset_master.xlsx - Netflix Movie Recommendation Ec.csv", header = None )
transactions = []

for i in range(0, 7466):
  transactions.append([str(data_set.values[i,j]) for j in range(0, 20)])
```

```python
print(data_set)
```

Apriopri training on data set

```python
from apyori import apriori
Basket_Intelligence = apriori(transactions = transactions, min_support=0.002, min_confidence = 0.2, min_lift = 3, min_length = 2)

```

Visualizing

```python
results =list(Basket_Intelligence)
results
```

Proper format

```python
def inspect(results):
  Movie1 = [tuple(result[2][0][0])[0] for result in results]
  Movie2 = [tuple(result[2][0][1])[0] for result in results]

  supports = [result[1] for result in results]
  return list(zip(Movie1, Movie2, supports))

DataFrame_Intelligence = pd.DataFrame(inspect(results), columns = ['Movie1', 'Movie2', 'support'])
```

```python
DataFrame_Intelligence
```

```python
DataFrame_Intelligence.nlargest(n=10, columns='support')
```
