Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing dataset

```python
data_set = pd.read_csv("/content/Dataset_master.xlsx - Big Basket.com Cart Apriori (Pr.csv", header = None )
transactions = []

for i in range(0, 7219):
  transactions.append([str(data_set.values[i,j]) for j in range(0, 20)])
```

```python
print(data_set)
```

Apriopri training on data set

```python
from apyori import apriori
Basket_Intelligence = apriori(transactions = transactions, min_support=0.002, min_confidence = 0.2, min_lift = 3, min_length = 2)

```

Visualizing

```python
results =list(Basket_Intelligence)
results
```

Proper format

```python
def inspect(results):
  product1 = [tuple(result[2][0][0])[0] for result in results]
  product2 = [tuple(result[2][0][1])[0] for result in results]

  supports = [result[1] for result in results]
  confidences = [result [2][0][2] for result in results]
  lifts = [result[2][0][3] for result in results]
  return list(zip(product1, product2, supports, confidences, lifts))

DataFrame_Intelligence = pd.DataFrame(inspect(results), columns = ['product1', 'product2', 'support', 'confidences', 'lifts'])
```

```python
DataFrame_Intelligence
```

```python
top_lift_rules = DataFrame_Intelligence.nlargest(n=10, columns='lifts')
print(top_lift_rules)
```
